# 8478
* 하루 당 외우는 단어수 A, 자면서 까먹는 단어수 B (A>B)
* 총 외어야 되는 단어수 N
* 몇일걸리는가?

```python
# N개를 외우는 순간 미션 종료 -> 밤에 까먹는 개수는 다음날 외우는 개수와 계산하기
# 즉, 첫날은 다 외운 단어로 간주하고 미리 계산해준다.
A,B,N = map(int,input().split())
N -= A
cap = A-B
days = 1 + N/cap
if(N%cap != 0): days += 1
print(int(days))
```

# 8480
* 거슬러주는 쿠폰의 개수 최소화
* 최대금액 쿠폰 먼저 거슬러주기 (greedy)

```python
n = int(input())
list = [50000,10000, 5000, 1000, 500, 100, 50, 10]
count = 0
for i in list :
    spare = n % i
    count += int(n / i)
    n = spare
print(count)
```

# 8481
* 사람이름과 IQ 지수가 주어질 때, IQ 상위 3명 이름 출력


```python
n = int(input())
people = []
for i in range(n):
    s, n = input().split()
    people.append((s,int(n)))
    
sorted_people = sorted(people, key = lambda x : x[1])
for i in range(3):
    print(sorted_people[-(i+1)][0])
    
```
