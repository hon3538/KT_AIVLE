## Zigbang 원룸 매물 데이터 수집

```python
import requests
import pandas as pd
```

#### Process
    - 동이름으로 위도 경도 구하기
    - 위도 경도로 geohash 알아내기
    - geohash로 매물 아이디 가져오기
    - 매물 아이디로 매물 정보 가져오기

### 1\. 동이름으로 위도 경도 구하기

```python
#'https://apis.zigbang.com/v2/search?leaseYn=N&q=%EC%84%B1%EC%88%98%EB%8F%99&serviceType=%EC%9B%90%EB%A3%B8'

# encode 된 한글을 decode 해서 넣어준다.
addr ='소사동'
url = f'https://apis.zigbang.com/v2/search?leaseYn=N&q={addr}&serviceType=원룸'
response = requests.get(url)
response
```

```python
response.json()
```

```python
# 일단 원룸 하나의 위도 경도 추출 -> 해쉬코드 알아내기 위함
data = response.json()['items'][0]
lat, lng = data['lat'], data['lng']
lat, lng
```

### 2\. 위도 경도로 geohash 알아내기

```python
geohash = geohash2.encode(lat, lng)
```

### 3\. geohash로 매물 아이디 가져오기

```python
url = f'https://apis.zigbang.com/v2/subways/261/items/oneroom?geohash={geohash}&radius=1&depositMin=0&rentMin=0&salesTypes[0]=전세&salesTypes[1]=월세&checkAnyItemWithoutFilter=true&domain=zigbang'
response = requests.get(url)
data = response.json()
data
```

```python
# id만 추출]
id_list = []
for d in data['items']:
    id_list.append(d['itemId'])
id_list
```

### 4\. 매물 아이디로 매물 정보 가져오기

```python
url ='https://apis.zigbang.com/v2/items/list'
params = {
    'domain' :'zigbang',
    'item_ids':id_list
}
response = requests.post(url, json=params)
response
```

```python
data = response.json()['items']
data
```

```python
# Dataframe
print(len(data))
df = pd.DataFrame(data)
df
```

```python
# 원하는  컬럼만 추출
df.columns
```

```python
columns = [ 'images_thumbnail', 'sales_type', 'deposit', 'rent', '전용면적',
       'room_type_title', 'floor', 'building_floor', 'title', 'status', 'tags', 'service_type',
        'manage_cost', 'reg_date', 'contract', 'address']
df = df[columns]
df.head(1)
```

```python
# 전용 면적 평수만 남기기
temp = []
for i in df['전용면적']:
    temp.append(float(i['p']))
temp
```

```python
df['전용면적'] = temp
```

```python
df.to_excel('zigbang1.xlsx', index=False)
```

```python
dir(pd.DataFrame) 
```

```python
dir(pd.Series) # iterrows는 df만 있음
```

```python
# 이미지로 저장
for idx, data in df.iterrows():
    width = 400
    height = 300
    link = data['images_thumbnail'] + f'?w={width}&h={height}&q=70&a=1'
    response = requests.get(link)
    
    with open(f'./image_data/test{idx}.png', 'wb') as file :
        file.write(response.content) 
```
