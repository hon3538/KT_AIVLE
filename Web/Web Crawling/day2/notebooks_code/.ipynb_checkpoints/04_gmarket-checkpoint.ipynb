### Gmarket
- 베스트 상품 200개 데이터 수집
- 상품의 이미지 200개 다운로드

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

#### 1. URL 찾기

```python
url = 'https://www.gmarket.co.kr/n/best'
```

#### 2. request > response : str(html)

```python
response = requests.get(url)
response
```

```python
response.text[:200]
```

#### 3. bs > DataFrame

```python
help(BeautifulSoup.__init__)
```

```python
dom = BeautifulSoup(response.text, 'html.parser')
dom
```

```python
selector = '.best-list li'
elements = dom.select(selector)
len(elements)
```

```python
elements[0]
```

```python
type(elements[0])
```

```python
# 제목, link, image 만 꺼내오기
# element = elements[0]
items = []
for element in elements:
    data = {
        'title' :element.select_one('.itemname').text,
        'link' : element.select_one('.itemname').get('href'),
        'img' : 'https:' + element.select_one('.thumb img').get('src'),
        'price' : int(element.select_one('.s-price').text.split(' ')[0][3:-1].replace(',',''))
    }
    items.append(data)
items
```

```python
items_df = pd.DataFrame(items)
items_df.head(2)
```

```python
items_df.to_excel('./gmarget_bestitems.xlsx', index=False, encoding='utf-8-sig')
```
