## DataFrame Merge

```python
# 판다스 패키지 로딩
import pandas as pd
```

### Inner Join

```python
# 데이터 merge 실습용 두 개의 데이터프레임 생성
data1 = {
    '학번':[1,2,3,4],
    '이름':['아이유','김연아','홍길동','강감찬'],
    '학과':['철학','경영학','컴퓨터','물리학']
}
data2 = {
    '학번':[1,2,4,5],
    '학년':[2,4,3,1],
    '학점':[1.5,2.0,4.1,3.8]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
# inner join -> 공유하는 컬럼이 학번이다. 일치되는 것만 merge
inner_df = pd.merge(df1, df2, on='학번', how='inner') # 어떤 컬럼을 기준으로 merge할 지 on 파라미터에 명시
display(inner_df)
```

### Outer Join

```python
# outer join -> 모든 데이터를 병합하고 빈자리는 결측처리한다.
outer_df = pd.merge(df1, df2, on='학번', how='outer')
display(outer_df)
```

### Left and Right

```python
left_df = pd.merge(df1, df2, on='학번', how='left')
display(left_df)
```

```python
right_df = pd.merge(df1, df2, on='학번', how='right')
display(right_df) # df2의 키를 기준으로 merge 진행
```

### Merge 응용 : 현실 데이터를 다룰 때 생길만 한 경우

#### 1) 같은 의미의 다른 컬럼 명을 병합할 때

```python
# 학번 == 학생고유번호
data1 = {
    '학번':[1,2,3,4],
    '이름':['아이유','김연아','홍길동','강감찬'],
    '학과':['철학','경영학','컴퓨터','물리학']
}
data2 = {
    '학생고유번호':[1,2,4,5],
    '학년':[2,4,3,1],
    '학점':[1.5,2.0,4.1,3.8]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)
```

```python
merged_df = pd.merge(df1, df2, left_on='학번', right_on='학생고유번호',how='inner')
display(merged_df)
```

```python
# 두 개의 공통 키중 하나를 삭제
merged_df.drop('학생고유번호', axis=1, inplace=True)
display(merged_df)
```

#### 2) 공유하는 컬럼이 없는 경우

```python
data1 = {
    '학번':[1,2,3,4],
    '이름':['아이유','김연아','홍길동','강감찬'],
    '학과':['철학','경영학','컴퓨터','물리학']
}
data2 = {
    '학년':[2,4,3,1],
    '학점':[1.5,2.0,4.1,3.8]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2, index=[1,2,4,5]) # 예제를 위해 이상한 인덱스 부여
display(df1)
display(df2)
```

```python
# 양 데이터프레임의 인덱스로 병합 진행
merged_df1 = pd.merge(df1, df2, left_index=True, right_index=True, how='inner')
display(merged_df1)
```

```python
# 왼쪾은 학번을 기준으로
# 오른쪽은 인덱스를 기준으로 병합
merged_df2 = pd.merge(df1, df2, left_on ='학번', right_index=True, how='inner')
display(merged_df2)
```

```python
# join 사용 : 양쪽 데이터프레임의 인덱스를 기준으로 병합
joined_df = df1.join(df2, how='inner') # on 은 default가 index이다.
display(joined_df)
```

### 데이터 연결 pandas.concat

```python
# 필요 라이브러리 로딩
import numpy as np
import pandas as pd

np.random.seed(1) # 예제를 위한 고정된 난수 설정
```

```python
# 3x2 정수로 구성된 매트릭스 생성
# np.random,randint(a, b, shape) -> a부터 b 사이의 random int를 shape 크기로 생성
df1 = pd.DataFrame(np.random.randint(0,9,(3,2)), index=['a','b','c'], columns=['one','two'])
display(df1)
```

```python
# 2x2 정수로 구성된 두 번째 매트릭스 생성
df2 = pd.DataFrame(np.random.randint(0,9,(2,2)), index=['a','b'], columns=['three','four'])
display(df2)
```

```python
# df1 & df2 concat으로 열방향 연결
concat_df = pd.concat([df1, df2], axis=1) # 연결할 dataframe을 리스트로 줘야한다.
display(concat_df)
```

```python
# outer join 으로도 concat과 같은 결과 데이터프레임을 확인할 수 있음
outer_df = pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
display(outer_df)
```

```python
# 행방향 연결
concat_df2 = pd.concat([df1, df2], axis=0)
display(concat_df2)
```

```python
# 새로운 데이터 생성
data1 = {
    '학번':[1,2,3,4],
    '이름':['아이유','김연아','홍길동','강감찬'],
    '학과':['철학','경영학','컴퓨터','물리학']
}
data2 = {
    '학번':[1,2,4,5],
    '이름':['김범수','이을용','이순신','김한국'],
    '학과':['경제학', '국문학', '경영학', '경영학']
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)
```

```python
# pandas concat 예시
concat_df = pd.concat([df1, df2], axis=0) # 행방향 연결
display(concat_df)
display(concat_df.reset_index()) # index를 순서대로 재할당하는데, 기존의 index를 컬럼으로 추가해버린다
display(concat_df.reset_index(drop=True)) # view 말고 원본에 바로 적용하려면 inplace = true
```
