```python
data = data.rename(columns={'버스정류장ARS번호':'ARS'})
```

```python
data.head()
bus_stop = data.copy()
```

```python
# ARS로 지역 분류
print(bus_stop.info())

temp = []
for st in bus_stop['ARS'] :
    temp.append(st[0:2])

temp = np.array(temp)
bus_stop['지역구'] = temp
bus_stop.head()
```

```python
# mapping
bus_stop['지역구'] = bus_stop['지역구'].map({
    '01': '종로구',
    '02': '중구',
    '03': '용산구',
    '04': '성동구',
    '05': '광진구',
    '06': '동대문구',
    '07': '중랑구',
    '08': '성북구',
    '09': '강북구',
    '10': '도봉구',
    '11': '노원구',
    '12': '은평구',
    '13': '서대문구',
    '14': '마포구',
    '15': '양천구',
    '16': '강서구',
    '17': '구로구',
    '18': '금천구',
    '19': '영등포구',
    '20': '동작구',
    '21': '관악구',
    '22': '서초구',
    '23': '강남구',
    '24': '송파구',
    '25': '강동구'})
bus_stop['지역구'].value_counts()
```

```python
bus_stop.shape
```

```python
# 결측치 제거
bus_stop = bus_stop.dropna()
# 컬럼별 결측치 확인
bus_stop.info()
```

```python
# ARS 다시 정수로 변환
bus_stop['ARS'] = bus_stop.loc[:,'ARS'].astype(int)
```

```python
# 자치구별 버스정류장 고유 개수 확인 후 변수로 저장
# 자치구별 groupby
bus_stop['bus_stop_count']=bus_stop.groupby('지역구', as_index=False).nunique()['ARS']
```

```python
only_seoul.info()
```

```python
only_seoul = only_seoul.rename(columns = {'ARS_ID':'ARS'})

# only_seoul과 bus_stop 병합
# inner 병합하면 서울 정류장 데이터만 남는다.
merged_data = pd.merge(bus_stop,only_seoul,on='ARS',how='inner')
merged_data.head()
```
