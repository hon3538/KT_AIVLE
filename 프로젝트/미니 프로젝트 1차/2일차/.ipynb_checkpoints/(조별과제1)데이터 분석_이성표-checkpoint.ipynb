## 3.이변량 분석

* 자료의 종류에 맞게 X --> Y 에 대해서 그래프(시각화)와 가설검정(수치화)를 수행하고 결과를 평가합니다.

* 가설검정시 다음의 항목을 참조하여 수행합니다.
    * 유의수준 : 5%
    * 숫자 --> 숫자 : 상관분석
    * 범주 --> 범주 : 카이제곱검정
    * 범주 --> 숫자 : t검정, 분산분석
    * 숫자 --> 범주 : 로지스틱 회귀모형을 통해, 회귀계수의 P.value로 검정을 수행합니다.
    
상관계수 (우리가 배운 기준)
- 0.5 ~ 1 : 강한
- 0.2 ~ 0.5 : 중간
- 0.1 ~ 0.2 : 약간
- 0 ~ 0.1 : 없음
- 음수 똑같음

p-value 기준을 0.05로 가져죠? 5%
p-value가 0.05보다 크다 == 귀무가설을 기각 하지 못한다 == 대립가설이 틀렸다(우리가 원하는 바는 대립가설)

- 귀무가설을 기각하려면 우리가 싫어하는 귀무가설을 입증할 증거가 나올 확률이 5% 이하여야 함
- p-value가 그 귀무가설을 입증할 증거가 나올 확률을 의미함

```python
# 각 자치구별 인구 분포 시각화
plt.figure(figsize=(10, 8))
sns.barplot(x='노선수', y='자치구', data=df.sort_values('노선수', ascending=False))
plt.title('자치구별 노선 수')
plt.xlabel('노선 수')
plt.ylabel('자치구')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.scatterplot(x='합계', y='총승객수', hue='자치구', data=df)
plt.title('자치구별 인구 대비 대중교통 이용객 수')
plt.xlabel('인구 수')
plt.ylabel('총승객수')
plt.legend(title='자치구', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sorted_df = df.sort_values('평균 이동 시간(분)', ascending=True)
plt.figure(figsize = (10,8))
sns.barplot(y='평균 이동 시간(분)', x='자치구', data=sorted_df)
plt.xticks(rotation = 30)
plt.title('도착 시군구별 평균 이동 시간')
plt.ylabel('평균 이동 시간(분)')
plt.xlabel('도착 시군구')
plt.show()
```

```python
df_melted = df.melt(id_vars='자치구', value_vars=['남자', '여자'], var_name='성별', value_name='인구수')

sns.pointplot(x='자치구', y='인구수', hue='성별', data=df_melted, dodge=True)
plt.xticks(rotation=45)
plt.title('자치구별 남성 및 여성 인구수 비교')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 자치구별 총 인구수
sorted_df = df.sort_values('합계', ascending=True)
sns.barplot(x='자치구', y='합계', data=sorted_df)
plt.xticks(rotation=45)
plt.title('자치구별 총 인구수')
plt.show()
```

```python

#자치구별 남녀 인구 비율 비교
plt.figure(figsize=(12, 8))
data_sorted = df.sort_values('합계', ascending=False)
bar_width = 0.35
index = np.arange(len(data_sorted))

bar1 = plt.bar(index, data_sorted['남자'], bar_width, label='남자')
bar2 = plt.bar(index + bar_width, data_sorted['여자'], bar_width, label='여자')

plt.xlabel('자치구')
plt.ylabel('인구 수')
plt.title('자치구별 남녀 인구 비율')
plt.xticks(index + bar_width / 2, data_sorted['자치구'], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# 자치구별 한식음식점 종사자수 대비 커피전문점 종사자수 비교
plt.figure(figsize=(12, 8))
bar_width = 0.35
index = np.arange(len(data_sorted))

bar1 = plt.bar(index, data_sorted['한식음식점_종사자수'], bar_width, label='한식음식점 종사자수', color='b')
bar2 = plt.bar(index + bar_width, data_sorted['커피전문점_종사자수'], bar_width, label='커피전문점 종사자수', color='r')

plt.xlabel('자치구')
plt.ylabel('종사자 수')
plt.title('자치구별 한식음식점 대비 커피전문점 종사자 수')
plt.xticks(index + bar_width / 2, data_sorted['자치구'], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# 자치구별 평균 이동 시간(분) 시각화
plt.figure(figsize=(12, 8))
sns.barplot(x='평균 이동 시간(분)', y='자치구', data=df.sort_values('평균 이동 시간(분)', ascending=False))
plt.title('자치구별 평균 이동 시간(분)')
plt.xlabel('평균 이동 시간(분)')
plt.ylabel('자치구')
plt.show()

# 자치구별 평균 이동인구 시각화
plt.figure(figsize=(12, 8))
sns.barplot(x='평균 이동인구', y='자치구', data=df.sort_values('평균 이동인구', ascending=False))
plt.title('자치구별 평균 이동인구')
plt.xlabel('평균 이동인구')
plt.ylabel('자치구')
plt.show()
```
